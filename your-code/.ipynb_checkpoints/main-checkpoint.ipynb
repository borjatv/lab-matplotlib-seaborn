#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
a = np.arange(0,10)
x2 = a**2
ex = np.e**a
```

```python
fig, ax = plt.subplots(1,2, figsize=(20,8))

ax[0].plot(x2, label='x^2', linewidth=2, color='b')
ax[0].plot(ex, label='e^x', linewidth=2, color='r')
#ax[0].set_xbound(upper=10, lower=-1)
#ax[0].set_ybound(upper=100, lower=-1)

ax[1].plot(x2, label='x^2', linewidth=2, color='b')
ax[1].plot(ex, label='e^x', linewidth=2, color='r')
ax[1].set_xscale('log')                          # Con esto fijamos la escala logarítmica en la 2ª gráfica
#ax[1].set_ybound(upper=10000, lower=-1)

ax[0].set_title('Normal scale', fontsize='xx-large') 
_ = ax[1].set_title('Log scale', fontsize='xx-large')
# ax[0].legend(loc="upper left", fontsize='x-large')
# _ = ax[1].legend(loc="upper left", fontsize='x-large')
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.

```python
df['Months_encoded'].value_counts('Months_encoded')
```

```python
# Febrero no debería tener 30 días, así que borramos duplicados por fecha
df.drop_duplicates('Date', inplace=True)
```

```python
# Otra opción: df.pivot_table(values='Steps', columns='Months_encoded', aggfunc='median').transpose()
step = df[['Steps', 'Months_encoded']].groupby('Months_encoded').median()
step
```

```python
cal = df[['Steps', 'Days_encoded']].groupby('Days_encoded').median()
cal
```

```python
wow = df[['Steps', 'Work_or_Weekend']].groupby('Work_or_Weekend').median()
wow
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fig, ax = plt.subplots(1,2, figsize=(20,8))
step.plot(kind='bar', xlabel='Months', ax=ax[0])
ax[0].set_xticks(range(12))
ax[0].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
ax[0].set_title('Median steps by month')

# ESTE GRÁFICO LO HABÍA HECHO APARTE PARA PROBAR

#cal.plot(kind='bar', xlabel='Days', ax=ax[1])
#ax[1].set_xticks(range(7))
#ax[1].set_xticklabels(['Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa', 'Su'])

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
wow.plot(kind='barh', xlabel='Work or weekend', ax=ax[1])
ax[1].set_yticks([0,1])
ax[1].set_yticklabels(['Weekend', 'Work'])
ax[1].set_xlim(5000,11000)
ax[1].set_title('Median steps by work or weekend')

fig.suptitle('Steps figures')
fig = plt.gcf()    # Sacamos el objeto figura

```

```python
# Lo guardamos

fig.savefig('figures.png')
```

```python
# ESTOS 2 LOS SAQUÉ APARTE PARA PRACTICAR plt.sca() Y plt.xticks() EN LUGAR DE set_xticks()

fig, ax = plt.subplots(1,2, figsize=(20,8))

plt.sca(ax[0])
plt.bar(x=step.index, height=step['Steps'], width=0.55)
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.xlabel('Months')

plt.sca(ax[1])
plt.bar(x=cal.index, height=cal['Steps'], width=0.55)
plt.xticks(ticks=range(7), labels=['Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa', 'Su'])
plt.xlabel('Days')
```
